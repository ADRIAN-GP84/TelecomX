```python
import json
import pandas as pd
from pandas import json_normalize

# Leer archivo JSON
with open('/content/TelecomX_Data.json') as f:
    datos_raw = json.load(f)

# Aplanar la estructura anidada
datos_clientes = json_normalize(datos_raw)

# Verifica las primeras filas
datos_clientes.head()
```

```python
print(datos_clientes.columns.tolist())
```

```python
# Eliminar los prefijos antes del punto en los nombres de columna
datos_clientes.columns = datos_clientes.columns.str.replace(r'^.*\.', '', regex=True)
```

```python
datos_clientes.columns.duplicated().any()
```

```python
datos_clientes.rename(columns={
    'customer.gender': 'gender',
    'customer.SeniorCitizen': 'senior',
    'customer.Partner': 'partner',
    'customer.Dependents': 'dependents',
    'customer.tenure': 'tenure',
    'phone.PhoneService': 'phone_service',
    'phone.MultipleLines': 'multiple_lines',
    'internet.InternetService': 'internet',
    'internet.OnlineSecurity': 'online_security',
    'internet.OnlineBackup': 'online_backup',
    'internet.DeviceProtection': 'device_protection',
    'internet.TechSupport': 'tech_support',
    'internet.StreamingTV': 'streaming_tv',
    'internet.StreamingMovies': 'streaming_movies',
    'account.Contract': 'contract',
    'account.PaperlessBilling': 'paperless_billing',
    'account.PaymentMethod': 'payment_method',
    'account.Charges.Monthly': 'monthly_charges',
    'account.Charges.Total': 'total_charges'
}, inplace=True)
```

```python
print(datos_clientes.columns.tolist())
```

```python
datos_clientes.head()
```

```python
datos_clientes.info()
```

```python
import pandas as pd

# Convertir la columna 'Total' a num√©rico
datos_clientes['Total'] = pd.to_numeric(datos_clientes['Total'], errors='coerce')
```

```python
# Contar valores nulos en 'Total' despu√©s de la conversi√≥n
faltantes_total = datos_clientes['Total'].isnull().sum()
print(f"Valores faltantes o no convertibles en 'Total': {faltantes_total}")
```

```python
# Mostrar filas con valores problem√°ticos en 'Total'
datos_clientes[datos_clientes['Total'].isnull()][['Total', 'customerID']]
```

```python
#Registros vacios o sin datos
datos_clientes['Total'].fillna(0, inplace=True)
```

```python
datos_clientes['Total'] = datos_clientes['Total'].fillna(0)
```

### **Visualizaciones**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot para ver c√≥mo var√≠a el total seg√∫n si el cliente se fue o no
sns.boxplot(data=datos_clientes, x='Churn', y='Total', palette='Set2')
plt.title("Total de cargos vs. Churn")
plt.ylabel("Cargos totales")
plt.show()
```

```python
promedios = datos_clientes.groupby('Churn')['Total'].mean()
print(promedios)
```

### **An√°lisis Exploratorio de Datos**

```python
print(datos_clientes['Churn'].value_counts(normalize=True) * 100)
```

```python
categorical_cols = datos_clientes.select_dtypes(include=['object', 'category']).columns.drop('Churn')

for col in categorical_cols:
    print(f"\nAn√°lisis para la columna: {col}")
    temp = pd.crosstab(datos_clientes[col], datos_clientes['Churn'], normalize='index') * 100
    print(temp)
```

```python
numerical_cols = datos_clientes.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_cols:
    print(f"\nEstad√≠sticas para la columna: {col}")
    print(datos_clientes.groupby('Churn')[col].describe()[['mean', '50%', 'std']])
```

### **Informe final**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(data=datos_clientes, x='Contract', hue='Churn', palette='Set1')
plt.title('Churn seg√∫n tipo de contrato')
plt.show()
```

```python
datos_clientes['Churn_num'] = datos_clientes['Churn'].map({'Yes':1, 'No':0})

correlations = datos_clientes[numerical_cols.tolist() + ['Churn_num']].corr()['Churn_num'].sort_values(ascending=False)
print("Correlaci√≥n de variables num√©ricas con churn:")
print(correlations)
```
